```python
import matplotlib.pyplot as plt
from scipy import stats

# Assuming your target column is 'price' in DataFrame 'df'
target_column = df['price']

# Create a histogram
plt.hist(target_column, bins=30, density=True, alpha=0.6, color='b', label='Histogram')

# Create a Q-Q plot
stats.probplot(target_column, dist="norm", plot=plt)
plt.title("Q-Q Plot and Histogram")
plt.legend()
plt.savefig('with_outliers.png')
plt.show()
```

```python
from scipy import stats

# Shapiro-Wilk test for normality
shapiro_stat, shapiro_p_value = stats.shapiro(target_column)

if shapiro_p_value < 0.05:
    print("The data is not normally distributed (reject null hypothesis).")
else:
    print("The data appears to be normally distributed (fail to reject null hypothesis).")
```

```python
skewness = target_column.skew()
kurtosis = target_column.kurtosis()

print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurtosis}")
```

```python
from scipy import stats

# Calculate the Z-scores for each data point
z_scores = np.abs(stats.zscore(target_column))

# Define a threshold for outlier detection (e.g., 2 or 3 standard deviations)
threshold = 3

# Identify and remove outliers
filtered_data = target_column[(z_scores < threshold)]

# Plot the histogram and Q-Q plot of the filtered data
plt.hist(filtered_data, bins=30, density=True, alpha=0.6, color='b', label='Histogram (Filtered)')
stats.probplot(filtered_data, dist="norm", plot=plt)
plt.title("Q-Q Plot and Histogram (Filtered)")
plt.legend()
plt.savefig('without_outliers.png')
plt.show()
```

```python
#Sample data (replace this with your actual data)
data = df["price"]

# Number of bins
num_bins = 4

# Calculate data range
data_min = data.min()
data_max = data.max()

# Calculate bin width
bin_width = (data_max - data_min) / num_bins

# Create bin edges
bin_edges = [data_min + i * bin_width for i in range(num_bins + 1)]

# Create labels for each bin based on the bin edges
bin_labels = [f"{int(bin_edges[i])}-{int(bin_edges[i+1])}" for i in range(num_bins)]

# Use pd.cut to assign labels to the bins and create a new column
df["PriceBin"] = pd.cut(data, bins=bin_edges, labels=bin_labels)

# Create a histogram
plt.hist(data, bins=num_bins, edgecolor='black')
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Histogram with Custom Bin Labels")
plt.show()

# Display the DataFrame with the new 'PriceBin' column
print(df)
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
from sklearn.preprocessing import label_binarize

# Assuming you have a multiclass classification model 'model'
# and 'X_test' is your test data

# Convert the multiclass labels to a binary matrix (one-hot encoding)
y_test_bin = label_binarize(y_test, classes=np.arange(5))

# Get the predicted probabilities for each class
y_score = model1.predict_proba(X_test)

# Initialize dictionaries to store ROC curve data for each class
fpr = {}
tpr = {}
roc_auc = {}

# Calculate ROC curve and AUC for each class
for i in range(5):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Compute micro-average ROC curve and ROC area
fpr["micro"], tpr["micro"], _ = roc_curve(y_test_bin.ravel(), y_score.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

# Plot ROC curves for each class
plt.figure(figsize=(10, 6))
for i in range(5):
    plt.plot(fpr[i], tpr[i], lw=2, label=f'Class {i} (AUC = {roc_auc[i]:.2f})')

# Plot the micro-average ROC curve
plt.plot(fpr["micro"], tpr["micro"],
         label='Micro-Average ROC (AUC = {:.2f})'.format(roc_auc["micro"]),
         color='navy', linestyle='--', linewidth=4)

# Add a random guessing line for reference
plt.plot([0, 1], [0, 1], 'k--', lw=2)

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Multiclass Classification (One-vs-Rest)')
plt.legend(loc="lower right")
plt.savefig('roc.png')
plt.show()
```
