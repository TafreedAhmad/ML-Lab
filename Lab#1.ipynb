```python
import pandas as pd

df = pd.read_csv('ecommerceData.csv')
df
```

Removed Empty Rows and Columns

```python
na_rows = df.isna().all(axis=1).sum()
print(f"Empty rows: {na_rows}")

# Number of columns with all NA values
na_cols = df.isna().all(axis=0).sum()
print(f"Empty columns: {na_cols}")
```

```python
# Remove empty rows
clean_data = df.dropna(axis=0, how='all')
print(f"Number of rows after removing empty rows: {clean_data.shape[0]}")

clean_data
```

```python

# Remove empty columns
clean_data = clean_data.dropna(axis=1, how='all')
print(f"Number of columns after removing empty columns: {clean_data.shape[1]}")
clean_data
```

```python
clean_data
```

Removing Duplicate Values

```python
# Calculate the number of duplicate entries
duplicate_count = clean_data.duplicated().sum()

# Print the result using string formatting
print(f"Duplicate entries: {duplicate_count}")
```

```python
# Determine the class of the 'item_id' column (data type)
column_class = clean_data['item_id'].dtype
print(f"Class of 'item_id' column: {column_class}")

# Calculate the number of unique values in the 'item_id' column
unique_count = clean_data['item_id'].nunique()
print(f"Number of unique values in 'item_id' column: {unique_count}")
```

```python
# Determine the class (df type) of the 'status' column
column_class = clean_data['status'].dtype
print(f"Class of 'status' column: {column_class}")

# Calculate the number of NA values in the 'status' column
na_count = clean_data['status'].isna().sum()
print(f"Number of NA values in 'status' column: {na_count}")

# Calculate the number of unique values in the 'status' column
unique_count = clean_data['status'].nunique()
print(f"Number of unique values in 'status' column: {unique_count}")

# List unique values in the 'status' column
unique_values = clean_data['status'].unique()
print(f"Unique values in 'status' column: {unique_values}")

# Count the number of rows for each order status
status_counts = clean_data['status'].value_counts()
print("Number of rows for each order status:")
print(status_counts)

# Replace 'received' and 'paid' with 'complete'
clean_data['status'].replace(['received', 'paid'], 'complete', inplace=True)

# Replace 'order_refunded' and 'refund' with 'refunded'
clean_data['status'].replace(['order_refunded', 'refund'], 'refunded', inplace=True)
```

```python
# Validate number of rows for each order status after replacing values
status_counts_after_replacement = clean_data['status'].value_counts()
print("Number of rows for each order status after replacing values:")
print(status_counts_after_replacement)
```

Finding Correlation

```python
numeric_data = clean_data.select_dtypes(include=['number'])
correlation_matrix = numeric_data.corr()
correlation_matrix
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a smaller heatmap with custom styling
plt.figure(figsize=(10,5))
sns.set_theme(style="white")

# Create the heatmap
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap="Blues", fmt='.1g')

plt.title("Correlation Heatmap")
plt.tight_layout()  # Ensure tight layout
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

# Count the number of orders for each category and sort them in descending order
category_counts = clean_data['category_name'].value_counts().reset_index()
category_counts.columns = ['category_name', 'n']
category_counts = category_counts.sort_values(by='n', ascending=False)

# Create a custom colormap with blues and light blues
colors = ['lightblue', 'lightblue', 'skyblue', 'skyblue', 'royalblue']
cmap = LinearSegmentedColormap.from_list('custom_cmap', colors)

# Create a bar plot of categories by their number of orders
plt.figure(figsize=(12, 6))  # Adjust the figure size
bars = plt.bar(category_counts['category_name'], category_counts['n'], color=cmap(category_counts.index))

# Highlight the column with the highest and lowest values
max_index = category_counts['n'].idxmax()
min_index = category_counts['n'].idxmin()
bars[max_index].set_color('royalblue')  # Highest value in royalblue
bars[min_index].set_color('red')  # Lowest value in lightblue

plt.xlabel('Category')
plt.ylabel('Number of orders')
plt.title('Category by Orders')
plt.xticks(rotation=90)

# Create a legend for the highlighted bars
from matplotlib.patches import Patch
from matplotlib.lines import Line2D
legend_elements = [Line2D([0], [0], color='royalblue', lw=4, label='Highest'),
                   Line2D([0], [0], color='lightblue', lw=4, label='Lowest')]
plt.legend(handles=legend_elements, title='Highlighted', loc='upper right')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate total revenue for each category and sort in descending order
category_revenue = clean_data.groupby('category_name').apply(lambda x: (x['price'] * x['qty_ordered']).sum())
category_revenue = category_revenue.sort_values(ascending=False).reset_index(name='total_revenue')

# Create a reversed color palette with dark blues to light blue gradient
colors = sns.color_palette("Blues_r", n_colors=len(category_revenue))

# Create a bar plot of categories by their revenue with reversed gradient colors
plt.figure(figsize=(12, 6))
ax = sns.barplot(x='category_name', y='total_revenue', data=category_revenue, palette=colors)
plt.xticks(rotation=90)
plt.title("Category by Revenue")
plt.xlabel("Category")
plt.ylabel("Total Revenue")
plt.tight_layout()

# Save the revenue chart as an image with a different name (e.g., 'revenue_chart_reverse_gradient.png')
plt.savefig('revenue_chart_reverse_gradient.png')

# Show the chart
plt.show()
```
